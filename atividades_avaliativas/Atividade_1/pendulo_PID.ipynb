```python
import numpy as np
import matplotlib.pyplot as plt

dt= 0.001
tf= 80
m=0.5
g=9.8
l=0.5
b=0.1

t= np.arange(0.0, tf, dt)

theta=np.zeros(np.size(t))
theta_p=np.zeros(np.size(t))
theta_pp=np.zeros(np.size(t))
u=np.zeros(np.size(t))

theta[0]=np.radians(30)
sp=np.radians(-30)*np.ones(np.size(t))


iE = 0 #Inicialização da parcela integral
dE = 0 #Inicialização da parcela derivativa
e_old = 0
kp = 55
ki = 9.7
kd = 13.12


for k in range(0, (np.size(t))-1):
    e = sp[k]-theta[k]
    iE = iE + e * dt
    dE = (e - e_old)/dt
    e_old = e
    u[k+1] = kp * e + ki * iE + kd * dE


    theta_pp[k+1]=-((m*g*np.sin(theta[k])/l))-(b*theta_p[k]/l)+u[k]
    theta_p[k+1] = theta_p[k] + theta_pp[k+1]*dt
    theta[k+1]=theta[k]+theta_p[k+1]*dt


p0=plt.plot(t,theta,label="Posição[Radianos]")
plt.plot(t,sp,label="Referência[Radianos]")
plt.legend()
plt.show()

p0=plt.plot(t,theta_p,label="Velocidade[Radianos/s] ")
plt.legend()
plt.show()

p0=plt.plot(t,theta_pp,label="Aceleração[Radianos/s^2] ")
plt.legend()
plt.show()
```
