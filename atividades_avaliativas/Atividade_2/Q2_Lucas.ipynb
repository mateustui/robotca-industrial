```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

######################################################################################
############################### FUNÇÕES DE ROTAÇÃO ###################################
######################################################################################

mat_rot_trans = np.array([[0.92718385, -0.37460659, 3],
                         [0.37460659,  0.92718385, 2],
                         [0         ,  0         , 0]])
  
######################################################################################
############################### APLICANDO AS ROTAÇÕES ################################
######################################################################################
print(mat_rot_trans)
# Vetores dados no sistema {A}
Oa = np.array([0,0,1]) #Origem de {A}
Xa = np.array([5,0,1]) # seta do eixo x (vermelho)
Ya = np.array([0,5,1]) # seta do eixo y (azul)



print(Oa)
Va = np.array((3,3, 1)) #seta do vetor (verde)

Vb = np.dot(mat_rot_trans,Va) #seta do vetor (verde)
print(Vb)
# Vetores convertidos para o sistema {B}
# Esta simplificação de código é válida, pois as coordenadas dos vetores unitários tanto em {A} como em {B} são as mesmas 
Ob = np.dot(mat_rot_trans,Oa) # Origem de {A} rotacionada para o sistema {B}
Xb = np.dot(mat_rot_trans,Xa) # seta do eixo x (vermelho), rotacionada para o sistema {B}
Yb = np.dot(mat_rot_trans,Ya) # seta do eixo y (azul), rotacionada para o sistema {B}

######################################################################################
############################### PLOT #################################################
######################################################################################
plt.figure()
ax = plt.gca()

# Plota o vetor [v]
ax.quiver(Oa[0], Oa[1], Va[0], Va[1], color='g', width = 0.01, angles='xy', scale_units='xy', scale=1)
ax.quiver(Ob[0], Ob[1], Vb[0], Vb[1], color='y', width = 0.01, angles='xy', scale_units='xy', scale=1)

ax.quiver(Oa[0], Oa[1], Xa[0], Xa[1], color='r', width = 0.01, angles='xy', scale_units='xy', scale=1)
ax.quiver(Oa[0], Oa[1], Ya[0], Ya[1], color='b', width = 0.01, angles='xy', scale_units='xy', scale=1)

ax.quiver(Ob[0], Ob[1], Xb[0], Xb[1], color='r', width = 0.01, angles='xy', scale_units='xy', scale=1)
ax.quiver(Ob[0], Ob[1], Yb[0], Yb[1], color='b', width = 0.01, angles='xy', scale_units='xy', scale=1)


lim = 10
ax.set_xlim(-1, lim+5)
ax.set_ylim(-1,lim)
ax.set_aspect('equal', 'box')

plt.grid(True)
plt.show()

print("Coordenadas do vetor no sistema {A}:")
print("[x]=[%f]\n[y]=[%f]" % (Va[0], Va[1]))

RVa = np.dot(mat_rot_trans.transpose(),Va)
print("\nCoordenadas do vetor no sistema {B}:")
print("[x]=[%f]\n[y]=[%f]" % (RVa[0], RVa[1]))

print("A rotação foi de %f graus"%np.degrees(np.arccos(mat_rot_trans[0,0])))
print("A rotação foi de %f graus"%np.degrees(np.arcsin(mat_rot_trans[1,0])))
print("O sistema foi deslocado %f unidades em x"%mat_rot_trans[0,2])
print("O sistema foi deslocado %f unidades em Y"%mat_rot_trans[1,2])
```
